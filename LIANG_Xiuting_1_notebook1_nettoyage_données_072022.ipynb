## df_eco (macro_economy2017)

source: FAO: https://www.fao.org/faostat/en/#data/MK

```python
df_eco = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/P9/macro_economy2017.csv')
df_eco.head()
```

### Suprim√© les colonnes non pertinents

```python
# Changer les noms de colonnes et suprim√© les colonnes non pertinents
df_eco = df_eco.rename(columns = {'Value': 'GDP par habitant(USD)',
                                 'Area Code (FAO)': 'Code zone', 'Area': 'Zone'})
df_eco = df_eco.drop(['Domain Code', 'Domain', 'Element Code', 'Element', 'Item Code',
                      'Item', 'Year Code', 'Year', 'Unit', 'Flag', 'Flag Description', 'Note'],axis = 1)
df_eco.head()
```

# Donn√©es FAO: la jointure de 'df_volailles', 'df_pop', 'df_eco'

## df_distance (Distance)


source: https://www.distancefromto.net/distance-from-france-country

```python
# V√©rifier les donn√©es
df_distance = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/P9/Distance.csv')
df_distance.head()
```

```python
# V√©rifier les donn√©es
df_distance['Zone'].describe()
```

```python
# V√©rifier les donn√©es
df_distance.info()
```

## La jointure des donn√©es open source

```python
# Faire la jointure
df = pd.merge(df_open2, df_distance, on = 'Zone', how = 'left')
df.describe()
```

## Traiter les valeurs manquants

```python
# Il y a valeur manquant pour prix de poulet, stabilit√© politique et distance.
# Parce qu'il est certain que l'information existe. Remplacer par les moyennes.
a = df['Chicken Fillets (1kg/USD)']
b = df['Political_Stability']
c = df['Distance km']

df['Chicken Fillets (1kg/USD)'] = df['Chicken Fillets (1kg/USD)'].fillna(np.mean(a))
df['Political_Stability'] = df['Political_Stability'].fillna(np.mean(b))
df['Distance km'] = df['Distance km'].fillna(np.mean(c))
df.describe()
```

# l'exportation de donn√©es

```python
# Renommer les noms de colonnes
df.index.name = 'Index'
df = df.rename( columns = {'Population(mille personnes)': 'Population',
                           'Disponibilit√© alimentaire (Kcal/personne/jour)(Kcal/personne/jour)':'Disponibilit√© alimentaire',
                           'Disponibilit√© de prot√©ines en quantit√© (g/personne/jour)(g/personne/jour)': 'Disponibilit√© de prot√©ines',
                           'GDP par habitant(USD)': 'PIB par habitant',
                           'Importations - Quantit√©(Milliers de tonnes)': 'Importations',
                           'Exportations - Quantit√©(Milliers de tonnes)': 'Exportations',
                           'Chicken Fillets (1kg/USD)': 'Prix poulet'
                          })
df.head()
```
